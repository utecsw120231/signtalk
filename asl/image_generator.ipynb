```python
import cv2 #opencv
import os
import time
import uuid
```

```python
labels=['hello','thanks','yes','no','iloveyou']
number_of_images=15
```

```python
for label in labels:
    !mkdir {label}
    cap=cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    for imgnum in range(number_of_images):
        ret,frame=cap.read()
        imagename=os.path.join(label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imagename,frame)
        cv2.imshow('frame',frame)
        time.sleep(2)
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
    cap.release()
```
